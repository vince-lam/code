```python
import numpy as np
import seaborn as sns #to improve visuals
import pandas as pd
%matplotlib inline
df1 = pd.read_csv('df1',index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
# to get histogram off a column
# can call matplotlib/seaborn arguments for style
df1['A'].hist(bins=30)
```

```python
# one method of creating histograms off a dataframe column
df1['A'].plot(kind='hist')
```

```python
# another method
df1['A'].plot.hist() # we will be using this method
```

```python
# area plot
df2.plot.area(alpha=0.5) # add alpha for transparency
```

```python
# bar plot
df2.plot.bar()
```

```python
# stacked bar plot
df2.plot.bar(stacked=True)
```

```python
# histogram
df1['A'].plot.hist(bins=50)
```

```python
# line plot
df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1) # need to specify x and y
```

```python
df1.plot.scatter(x='A',y='B',c='C', cmap='coolwarm') # can change colour by a third argument
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

```python
# box plot
df2.plot.box()
```

```python
# hexagonal bin plot
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.head()
```

```python
# like a scatter plot but with hexbins
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```

```python
#kernal density estimation
df2['a'].plot.density()
```

```python
df2.plot.density()
```
