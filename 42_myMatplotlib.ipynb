```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

```python
# 2 ways to create a matplotlib plots
# 1st way via functional method 
# 2nd way via object orientated method (usually better method)
```

```python
#FUNCTIONAL
plt.plot(x,y)
# how to change x & y labels
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# how to put multi plots in the same canvas
# this specifices the number of rows and columns
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT ORIENTATED METHOD, more control
fig = plt.figure()

#([left of axis, bottom of axis, width of axis, heigh of axis])
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure() # more manual

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
# how to do subplots
# subplots allow you to specify the number of rows and columns
fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

#axes.plot(x,y)
# how to prevent overlapping
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()
```

```python
# how to configure figure size and DPI
fig = plt.figure(figsize=(8,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# how to save a figure
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

#ax.set_title('Title')
#ax.set_ylabel('Y')
#ax.set_xlabel('X')

# how to add legends
# loc=0 for best location
ax.legend(loc=0)
```

```python
# how to set colours
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y,color='green')
#ax.plot(x,y,color='#FF8C00') #can be more precise and use rgb hexcodes
# how to add line width too
ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=0.5,linestyle='--',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') #default linewidth is 1, alpha controls the transparency of the line
# can use lw and ls for shorthand
```

```python
# control over axes appearance
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',linewidth=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
